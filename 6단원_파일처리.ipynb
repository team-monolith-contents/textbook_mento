### **1) 파일 처리의 개요**

파일은 의미 있는 정보를 담기 위한 논리적 단위로, 프로그램에서 필수적인 존재이다. 여러 저장 매체(하드

연하게 정리하고 관리하기 위해 일반적으로 디렉터리를 만들어 그 안에 파일을 저장한다. 우리가 일반적으
로 파일을 다루는 방법을 연상하면서 파이썬에서 파일을 처리하는 원리에 대하여 알아보자. 

```python
import os
data = os.listdir(".")
print(data)
for d in data:
    print(d)
    print("is directory? : " + str(os.path.isdir(d)))
    print("is file? : " + str(os.path.isfile(d)))
```

### **1) 텍스트 데이터 읽기**

이제 생성된 파일을 불러 와 사용하는 방법을 알아보자. 이전 단원에서는 텍스트를 저장하기 위해 open 
함수를 사용하여 파일을 생성하고 write 메소드를 이용하여 텍스트를 저장했다. 이제 open 함수를 사용하
여 파일을 열어 내용을 읽은 후 저장된 텍스트를 출력해 보자. 

```python
f = open("text.txt", "w", encoding="utf8")

f.write("안녕하세요.")
f.write("\n")
f.write("반갑습니다.")
f.write("\n")

f.close()

# 파일 읽기
f = open("text.txt", "r", encoding="utf8")

str = f.read()
print(str)

f.close()
```

```python
f = open("text.txt", "r", encoding="utf8")

line1 = f.readline()
print("line 1 : " + line1)
line2 = f.readline()
print("line 2 : " + line2)
line3 = f.readline()
print("line 3 : " + line3)

f.close()
```

```python
import xml.etree.ElementTree as ET
import codecs
f = codecs.open("movie.xml", encoding="UTF-8")
str = f.read()
tree = ET.ElementTree(ET.fromstring(str))
root = tree.getroot()
print(root.tag)

for child in root:
    print("tag : " + child.tag + " : " + child.text)
f.close()
```

### **2) 이진 데이터 읽기**

이진 데이터를 저장하는 형식은 데이터의 종류에 따라 조금씩 달라질 수 있다. 하지만 저장된 데이터의 종
류와 상관없이 이진 데이터를 읽어 올 수 있어야 저장된 바이트의 의미를 해석할 수 있을 것이다. 저장된 
파일의 바이트 내용을 리스트 형태로 읽어 와 각 바이트에 어떤 값이 저장되어 있는지 조사하는 방법을 
알아보자. 

```python
import array
f = open("data.bin", "wb")

byte_arr = []
for i in range(0, 256):
    byte_arr.append(i)

data = bytes(byte_arr)

f.write(data)

f.close()

f = open("data.bin", "rb")
data = array.array("B")  # "B" 값을 전달하여 최소 0부터 255까지 숫자를 저장할 수 있는 리스트를 선언한다.
data.fromfile(f, 10)  # 이진 파일의 내용을 불러오기 위하여 fromfile 메소드를 사용한다. 두번째 인자는 읽어올 바이트의 개수. 

for item in data:
    print(item, end=", ")
f.close()
```

```python
import pickle
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return ("이름 : " + self.name + "\n나이 : " + self.age)
        
p1 = Person("철수", '20')
p2 = Person("영희", '24')
p3 = Person("주영", '32')
persons = [p1, p2, p3]

f = open("person_data.bin", "wb")
pickle.dump(p1, f)
f.close()

f = open("persons_data.bin", "wb")
pickle.dump(persons, f)
f.close()
```

```python
f = open("person_data.bin", "rb")
person = pickle.load(f)
print(person)

f.close()

f = open("persons_data.bin", "rb")
persons = pickle.load(f)

for p in persons:
    print(p)

f.close()
```
