### **1) 기초 예외 처리**

뜻하지 않은 예외 상황이 발생했을 때 이를 처리할 수 있도록 도와주는 try–except–finally 구문을 활용하는 
방법을 알아보자.

```python
arr = [1, 2, 3]
try:
    print(arr[2])
    print(arr[3])
except:
    print("리스트의 요소에 접근하지 못했습니다.")
```

```python
f = open("file.txt", "w")
try:
    f.write("Hello World")
    f.write(100)
except TypeError:
    print("타입 예외 발생(100은 쓸 수 없음)")
finally:
    print("예외 여부와 상관없이 무조건 실행")
    f.close()
```

### **2) 다양한 방식의 예외 처리**

try–except 블록을 사용하는 아주 기초적인 예외 처리 방법 이외에도 파이썬에서는 다양한 방식으로 예
외를 처리하는 방법을 제공한다. 예외 처리의 여러 방식을 살펴보자. 

```python
arr = [1, 2, 3]
try:
    print(arr[4])
except IndexError as e:
    print(e)
```

```python
arr = [1, 2, 3]
try:
    print(arr[0])
    print(arr[1])
    print(arr[2])
except:
    print("예외 발생!")
else:
    print("성공적으로 모든 코드를 실행!")
```

### **3) 사용자 정의 예외**

프로그램을 작성하다 보면 프로그램의 성격에 따라 사용자가 예외 클래스를 직접 작성할 필요가 있다. 이러한 
사용자 정의 예외를 작성하는 방법을 살펴보자

```python
class Error(Exception):
    pass
```

```python
class MyError(Exception):
    def __init__(self, message):
        self.message = message
    def __str__(self):
        return self.message
```

```python
class NotEvenNumberError(Exception):
    pass
def even_number_add(a, b):
    if a % 2 != 0 or b % 2 != 0:
        raise NotEvenNumberError
    return a + b
try:
    a = even_number_add(2, 2)
    print("결과 값 : " + str(a))
    b = even_number_add(4, 2)
    print("결과 값 : " + str(b))
    c = even_number_add(2, 1)
except NotEvenNumberError as e:
    print("짝수만 더할 수 있습니다.")
```
