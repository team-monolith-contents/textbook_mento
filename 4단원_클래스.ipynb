### **1) 클래스 구성**

클래스는 크게 변수 부분과 함수 정의 부분으로 구성된다. 클래스 내부에 변수를 정의하여 객체와 관련 있는 
정보를 저장할 수 있고, 함수를 작성하여 객체가 할 수 있는 행동을 정의할 수 있다. 이렇게 클래스와 관련된 
동작을 표현하기 위해 클래스 내부에 작성한 함수를 메소드라고 한다. 클래스의 구성에 대해 알아보자. 

````python
class Car:
    def __init__(self, type, speed):
        self.type = type
        self.speed = speed

c = Car("스포츠카", 100)
````

````python
class Car:
    def __init__(self, type, speed):
        self.type = type
        self.speed = speed

    def move(self):
        print(self.type+'가'+str(self.speed)+' 속도로 움직입니다.')

    def speed_up(self, amount):
        self.speed += amount

    def speed_down(self, amount):
        self.speed -= amount

c = Car("스포츠카", 100)
c.speed_up(10)
c.move()
c.speed_down(5)
c.move()
````

특이하게도 클래스에 정의된 모든 메소드의 첫 번째 파라미터로 self 값을 받는 것을 확인할 수 있다. 메소드가 호출될 때 메소드 내부에서 self 값을 통해 내부 변수 값에 접근할 수 있다. 그러나 실제로 메소드를 호출할 때는 self값을 직접 넘겨줄 필요가 없다. self 값은 자동으로 전달되므로 메소드 내부에서는 해당 값을 이용만 하면 된다. self 값에 접근하여 해당 객체에 존재하는 멤버 변수나 메소드에 접근할 수 있다.

### **2) 클래스 변수와 클래스 메소드**

모든 객체가 공유해야 할 공통 변수나 메소드를 선언해야 할 필요성이 있을 수 있다. 가령 현재까지 생성된 자
동차 객체의 개수를 구해야 하거나 self의 값 참조가 필요 없는 메소드를 선언해야 할 필요성이 생길 수 있다. 
파이썬에서는 모든 객체가 공유하는 클래스 변수와 클래스 이름을 통해 접근할 수 있는 클래스 메소드를 정의
하는 방법을 제공한다. 클래스 변수와 클래스 메소드에 대해 알아보자. 

````python
class Car:
    count = 0
    def __init__(self, type, speed):
        self.type = type
        self.speed = speed
        Car.count += 1

    @classmethod
    def get_count(cls):
        return cls.count


print(Car.get_count())
c1 = Car('스포츠카', 100)
c2 = Car('트럭', 50)
print(Car.get_count())
````

### **3) 클래스의 상속**

이미 만들어진 클래스를 확장하는 방식으로 클래스를 정의하고 싶은 경우가 있다. 자동차라는 클래스를 만
든 후 그 자동차의 속성을 그대로 이용하는 트럭이나 택시 클래스를 만들고자 하는 것을 예로 들 수 있다. 
이 경우 트럭이나 택시 모두 자동차이므로 기존의 자동차에 존재하는 변수와 메소드를 물려받고 동시에 트
럭이나 택시에서 필요한 새로운 변수와 메소드만 정의하는 것이 가능하다. 이를 상속이라고 한다. 클래스의 
상속에 대해 알아보자. 

  
  
```python
def __init__(self):
    super().__init__()
```

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def eat(self, food):
        print(self.name +'가 '+food+'를 먹습니다.')
````

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def eat(self, food):
        print(self.name +'가 '+food+'를 먹습니다.')

    def __str__(self):
        return self.name+'는 '+str(self.age)+'살입니다.'

class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary

    def work(self):
        print('열심히 일을 합니다.')

    def get_salary(self):
        print('급료를 받습니다.')
        return self.salary

e = Employee('영희', 19, 100)
print(e)
e.eat('밥')
e.work()
r = e.get_salary()
print('급료는 '+str(r)+'만 원입니다.')
````

### **4) 특수 메소드**

파이썬에서는 특수한 행동을 위해 예약된 이름의 특수 메소드들이 있다. 가령 __init__와 같은 생성자 메소
드도 객체에 필요한 정보의 초기화라는 특수한 행동을 정의하므로 특수 메소드에 포함된다. 이처럼 특별한 
용도로 사용될 수 있어 재정의하면 좋을 만한 메소드가 무엇이 있는지 살펴보자. 단, 특수 메소드는 반드시 
재정의해야 할 필요는 없고 상황에 따라 필요할 경우에만 재정의하면 된다.

````python
class DeletableClass:
    def __del__(self):
        print("delete")
        
d = DeletableClass()
del d
````

````python
class Person:
    def __init__(self, name, age, height):
        self.name = name
        self.age = age
        self.height = height
    def __str__(self):
        return "Person 설명, 이름은 " + self.name + " 키는 " + str(self.height)
    def __len__(self):
        return self.height

    def __getitem__(self, key):
        if key == "name":
            return self.name
        if key == "age":
            return self.age
        return None
        
p = Person("철수", 18, 170)
print(p)
print(len(p))
print(p['age'])
print(p['unknown'])
````
