### **1) 정수형**


**수식 연산자**  
숫자형 데이터는 정수, 실수, 복소수로 나눌 수 있다. 숫자형 데이터의 연산자는 다음과 같다.  
    

정수 계산은 셀 안에서 바로 실행될 수 있으나, 변수에 데이터를 저장해서 계산할 수도 있다.

<br>

**변수**  
- 정의: 데이터를 담는 장소이다.  
  데이터를 나중에 사용할 때 편리하도록 의미 있는 단어로 이름을 짓는다.  
  파이썬은 특별히 정수형이라고 자료형을 미리 선언하지 않고 바로 변수에 값을 대입해서 사용한다.  
  (동적 타이핑: 실행 중에 동적으로 자료형을 결정함)
- 파이썬의 변수 이름 규칙
> - 예약어는 사용할 수 없다
> - 첫번째 문자에는 숫자가 올 수 없다.
> - 공백을 사용할 수 없으므로 snake_case 또는 camelCase 형태로 사용한다. (변수와 함수 이름은 snake_case 스타일로, 클래스는 camelCase로 하는 것을 권장)
> - 밑줄 문자(_)는 사용할 수 있다.
> - 대문자, 소문자를 구별한다(a와 A는 다른 변수임)

```python
# 셀에서 바로 정수 계산하기
18+2
```

```python
# 정수 데이터를 변수에 저장하여 계산하기
age = 18
print(age)

after_2 = 2
print(age+after_2)
```

**대입 연산자**  
파이썬 문법에는 대입 연산자가 있다.    
할당 연산자는 연산을 수행한 후 연산 결과를 좌변의 변수에 할당해주는 연산자이다.  
 
 이때 좌변의 변수는 할당 연산자로 연산하기 전에 미리 선언해야 한다.    



```python
age = 18
age += 2
age -= 1
age += 1
age //= 2 
age **= 3

print(age) # 셀을 실행하기 전 출력 결과를 예상해보세요.
```

- 정수를 표현하는 여러 가지 방법

```python
print("{:d}".format(515)) #정수를 넣는다.

print("{:5d}".format(515)) #최소 5칸을 차지하고 정수를 넣는다.

print("{:+5d}".format(515)) #양수면 +를 표시한다.

print("{:=+5d}".format(515)) #+를 맨 앞에 표시한다.

print("{:05d}".format(515)) #빈칸은 0으로 채운다.

print("{:+05d}".format(515)) #양수면 0 앞에 +를 표시한다.
```

- 실수를 표현하는 여러 가지 방법

```python
print("{:f}".format(11.17))       #실수를 넣는다.

print("{:12f}".format(11.17))     #최소 12칸을 차지하고 실수를 넣는다.

print("{:12.1f}".format(11.17))   #소수점 1자리까지 반올림해서 나타낸다.
```

### **2) 튜플**

튜플은 복수의 요소로 구성된 하나의 그룹을 뜻한다.    
다양한 데이터를 하나로 묶어 주는 기능으로, 리스트와 비슷하지만 변경이 불가능하다.   
그 대신 생성 과정이 간단하고, 데이터의 변경이 없으므로 복사하지 않고 원본을 그대로 사용한다.   
튜플을 생성해 보자.

  
여러 개의 값을 담고 있지만 하나의 그룹으로 다룬다.  


```python
t = () #빈 튜플
print(type(t))
t
```

```python
xy = (2560, 1440) #(2560, 1440)
print(type(xy))
xy
```

```python
color = 129, 247, 216 #(129, 247, 216)
print(type(color))
color
```

```python
xy + color #(2560, 1440, 129, 247, 216)
```

```python
xy * 2 #(2560, 1440, 2560, 1440)
```

  
괄호를 생략해도 하나로 묶어서 대입되고, 여러 개로 대입할 때도 하나씩 들어간다.  


```python
color = 129, 247, 216   #패킹(129, 247, 216)
red, green, blue = color #언패킹

print(red)
print(green)
print(blue)
```

### **2) 연산자**

True나 False를 그대로 대입해 사용하는 것보다는 비교 연산자를 사용하는 것이 일반적이다.   
또는 여러 개의 비교 연산자의 관계식을 조합하는 논리 연산자를 사용한다.   
특히 비교 연산자는 나중에 배울 제어문과 반복문에 많이 쓰인다.   
연산자에 대해 알아보자. 


   
비교 연산자의 결과는 True 또는 False이다. 따라서 특정 조건에 맞는지 확인할 수 있다.   


  
여러 개의 True와 False를 조합하여 다음과 같이 논리 계산을 할 수 있다.     
|연산자|설명|
|-|-|
|A and B|A와 B가 모두 True일 때만 True|
|A or B|A와 B가 모두 False일 때만 False|
|not A|A가 True이면 False, A가 False이면 True|

```python
A = True
B = False

print('not A : ', not A)
print('A and B : ', A and B)
print('A and not B : ', A and not B)
print('A or B : ', A or B)
print('not A or B : ', not A or B)
```

   
컴퓨터는 모든 값을 0 또는 1 즉 비트의 모음으로 다룬다. 비트를 다루는 연산자는 다음과 같다.   
여기서는 bin() 함수를 이용해 이진수로 표현한다.   


```python
 bin(0b1010 & 0b0110)  
```

```python
 bin(0b1010 | 0b0110)  
```

```python
 bin(0b1010 ^ 0b0110)  
```

```python
 bin(~0b1010)  
```

```python
 bin(0b1010 << 2)
```

```python
 bin(0b1010 >> 2)
```
